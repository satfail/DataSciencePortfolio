## Histogramas


```python
cols = ["Median" , "ShareWomen", "Unemployment_rate"]

fig = plt.figure(figsize=(5,20))
for r in range(1,4):
    ax = fig.add_subplot(3,1,r)
    ax = recent_grads[cols[r-1]].plot(kind='hist', rot=45)
    ax.set_title(cols[r-1])
    
plt.subplots_adjust(hspace=1)    
plt.show()
```

Como podemos ver los histogramas son una buena herramienta para determinar la frecuencia con la que los valores aparecen en las columnas que hemos seleccionadp y su distribución.


## ScatterMatrix

```python
from pandas.plotting import scatter_matrix
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(6,6))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```

En los plots anteriores se ha creado una matriz en la cual la diagonal principal está formada por los histogramas de las variables que hemos determinado. Para cada fila tenemos la relación de cada variable con el resto. Por ejemplo:

Posición (1,3) = Sample_size vs Unemployement_rate
Posición (3,1) = Unemployement_rate vs Sample_size

Veáse que se trata de la mismo scatter con los ejes invertidos.
